```python
# Copying Dataset
AmazonSal_AnDf = AmazonSales_Df.copy()
```

```python
AmazonSal_AnDf.dropna(subset=['Discount Amount','Sales Price', 'Item Number'],inplace=True)
```

```python
AmazonSal_AnDf['Item Class'].value_counts()
```

```python
AmazonSal_AnDf.groupby('Item').sum()['Sales Quantity']
```

Resume on 21 Sept 2022

```python
# Creating Month_Year, Year, Month, Quarter,Day Columns 
AmazonSal_AnDf['Billed_Year'] = AmazonSal_AnDf['Invoice Date'].dt.year
AmazonSal_AnDf['Billed_Month_Year'] = AmazonSal_AnDf['Invoice Date'].dt.strftime('%y-%m')
AmazonSal_AnDf['Billed_Month'] = AmazonSal_AnDf['Invoice Date'].dt.month
AmazonSal_AnDf['Billed_Quater'] = AmazonSal_AnDf['Invoice Date'].dt.quarter
AmazonSal_AnDf['Billed_Day'] = AmazonSal_AnDf['Invoice Date'].dt.day
```

```python
# Only Required Columns to display
AmazonSal_AnDf = AmazonSal_AnDf[['CustKey','Item','Invoice Date','Billed_Year','Billed_Month_Year','Billed_Quater', 'Billed_Month','Billed_Day','Sales Quantity', 'Sales Amount Based on List Price','Discount Amount','Sales Amount', 'Sales Margin Amount','Sales Cost Amount','Sales Rep','U/M','List Price','Sales Price']]
```

```python
AmazonSal_AnDf.isnull().sum()
```

```python
AmazonSal_AnDf.head(10)
```

```python
AmazonSal_AnDf.tail(10)
```

```python
OverallYearly_Sales = AmazonSal_AnDf.groupby('Billed_Year').sum()['Sales Amount'].reset_index()
```

```python
OverallYearly_Sales[['Billed_Year', 'Sales Amount']]
```

```python
plt.figure(figsize=(15,7))
sns.barplot( y='Sales Amount',x='Billed_Year', data=OverallYearly_Sales, palette='gist_earth_r')
plt.title('Overall Yearly Sales', fontsize=15)
plt.xlabel('Years', fontsize=15)
plt.ylabel('Sales Amount', fontsize=15)
plt.show()
```

```python
# Piechart Share of  3 years sale
plt.figure(figsize=(20,6))
plt.pie('Sales Amount',labels='Billed_Year',data=OverallYearly_Sales,autopct='%1.2f%%')
plt.title('Yearly Sales Contribution', fontsize=15)
plt.legend(round(OverallYearly_Sales['Sales Amount'],2),loc=7,fontsize = 10)
plt.show()
```

```python
# Profit on Year
OverallYearly_Profi = AmazonSal_AnDf.groupby('Billed_Year').sum()['Sales Margin Amount'].reset_index()
```

```python
OverallYearly_Profi[['Billed_Year', 'Sales Margin Amount']]
```

```python
plt.figure(figsize=(15,7))
sns.barplot( x='Billed_Year',y='Sales Margin Amount', data=OverallYearly_Profi, palette='gist_earth_r')
plt.title('Overall Yearly Profits')
plt.xlabel('Years')
plt.ylabel('Sales Margin Amount')
plt.show()
```

```python
# Piechart Share of  3 years profit
plt.figure(figsize=(17,6))
plt.pie('Sales Margin Amount',labels='Billed_Year',data=OverallYearly_Profi,autopct='%1.2f%%',startangle=90)
plt.title('Yearly Profit Contribution', fontsize=15)
plt.legend(round(OverallYearly_Profi['Sales Margin Amount'],2),loc=7,fontsize = 8)
plt.show()
```
